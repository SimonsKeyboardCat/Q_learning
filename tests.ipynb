<h1>Tests for Q learning algorithm</h1>

```python
from frozen_lake import run_frozen_lake
import matplotlib.pyplot as plt
```

```python
# Parameters
seeds = [0 , 12, 48, 90, 1402]
alphas = [0.9, 0.7, 0.5, 0.3, 0.1]
episodes = 15000
```

```python
for seed in seeds:
   plt.clf()
   for alpha in alphas:
      name =  "seed" + str(seed) + "alpha" + str(alpha)
      legend_label = "alpha = " + str(alpha)
      run_frozen_lake(
         name_of_save_file=name, 
         episodes=episodes,
         seed=seed,
         legend_label=legend_label,
         learning_rate_a=alpha,
         discount_factor_g=0.9,
         epsilon=1,
         slippery=True,
         strategy="greedy", 
         is_training=True, 
         render=False)
      plt.legend(loc="upper left")   
   
```
