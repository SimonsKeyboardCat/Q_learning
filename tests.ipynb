<h1>Tests for Q learning algorithm</h1>

```python
from frozen_lake import run_frozen_lake
import matplotlib.pyplot as plt
```

```python
# Parameters
seeds = [0 , 12, 48, 90, 1402]
alphas = [0.9, 0.7, 0.5, 0.3, 0.1]
discounts = [0.9, 0.7, 0.5, 0.3, 0.1]
epsilons = [1, 0.98, 0.97, 0.95, 0.93]
taus = [0.2, 0.5, 1, 4, 10]
episodes = 15000
```

```python
for seed in seeds:
   plt.clf()
   for alpha in alphas:
      name =  "seed" + str(seed) + "alpha" + str(alpha)
      run_frozen_lake(
         name_of_save_file=name, 
         episodes=episodes,
         seed=seed,
         learning_rate_a=alpha,
         discount_factor_g=0.9,
         epsilon=1,
         slippery=True,
         strategy="greedy", 
         is_training=True, 
         render=False)
      plt.gca().legend(('0.9', '0.7', '0.5', '0.3', '0.1'))
      plt.savefig(name + ".png")
       
   
```

```python
for seed in seeds:
   plt.clf()
   for discount in discounts:
      name =  "seed" + str(seed) + "discount" + str(discount)
      run_frozen_lake(
         name_of_save_file=name, 
         episodes=episodes,
         seed=seed,
         learning_rate_a=0.5,
         discount_factor_g=discount,
         epsilon=1,
         slippery=True,
         strategy="greedy", 
         is_training=True, 
         render=False)
      plt.gca().legend(('0.9', '0.7', '0.5', '0.3', '0.1'))
      plt.savefig(name + ".png")
```

```python
for seed in seeds:
   plt.clf()
   for epsilon in epsilons:
      name =  "seed" + str(seed) + "greedy" + str(epsilon)
      run_frozen_lake(
         name_of_save_file=name, 
         episodes=episodes,
         seed=seed,
         learning_rate_a=0.9,
         discount_factor_g=0.9,
         epsilon=epsilon,
         slippery=True,
         strategy="greedy", 
         is_training=True, 
         render=False)
      plt.gca().legend(('1', '0.98', '0.97', '0.95', '0.93'))
      plt.savefig(name + ".png")
```

```python
for seed in seeds:
   plt.clf()
   for tau in taus:
      name =  "seed" + str(seed) + "boltzmann" + str(tau)
      run_frozen_lake(
         name_of_save_file=name, 
         episodes=episodes,
         seed=seed,
         learning_rate_a=0.9,
         discount_factor_g=0.9,
         epsilon=1,
         slippery=True,
         strategy="boltzman", 
         is_training=True, 
         render=False)
      plt.gca().legend(('0.2', '0.5', '1', '4', '10'))
      plt.savefig(name + ".png")
```

```python
from frozen_lake_tests import run_frozen_lake_test
import numpy as np
```

```python
name = "seed0alpha"

for alpha in alphas:
   last_episodes = []
   last_episodes.append(
      run_frozen_lake_test(
         name_of_save_file=name + str(alpha), 
         episodes=15000,
         slippery=True,
         strategy="greedy",
         render=False
      )
   )
   print("Alpha: ", alpha)
   print(last_episodes)
   mean = np.mean(last_episodes)
   s2 = np.var(last_episodes)
   s = np.std(last_episodes)
   print("Mean: ", mean)
   print("Variance: ", s2)
   print("Standard deviation: ", s)
   print("")


```
